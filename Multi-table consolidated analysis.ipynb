```python
import datetime
def get_time(list1,list2):

    temp_date = []
    for i, line in enumerate(list2):
        try:
            if "-" in line:
                a = datetime.datetime.strptime(list1[i],'%Y-%m-%d') - datetime.datetime.strptime(line,'%Y-%m-%d')
                temp_date.append(a.days)
            else:
                temp_date.append(-1)
        except:
            temp_date.append(-1)
    return temp_date
```

```python
total_pd = total_pd.sort_values(by=["患者编号","访视时间"])
first_times = []
id_temp = list(set(total_pd["患者编号"].values))
id_temp.sort()
timess = []
for line in id_temp:
    
    tt = total_pd[total_pd["患者编号"] == line]["访视时间"].values
    for i, t in enumerate(tt[:]):
        if i == 0:
            first_times.append(0)
        else:
            first_times.append(t - tt[0])
        if i != len(tt) - 1:
            timess.append(tt[-1] - t)
    timess.append(0)
total_pd["访视时间"] = first_times
del first_times
del timess
```
