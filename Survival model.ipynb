```python
import numpy as np

def fit_and_score_features(X, y):
    n_features = X.shape[1]
    scores = np.empty(n_features)
    m = CoxPHSurvivalAnalysis()
    for j in range(n_features):
        Xj = X[:, j:j+1]
        m.fit(Xj, y)
        scores[j] = m.score(Xj, y)
    return scores

scores = fit_and_score_features(np.concatenate([tt[train_feature].values[:39],tt[train_feature].values[40:index]],axis=0), np.concatenate([new_data_y[:39],new_data_y[40:index]],axis=0))
pd.Series(scores, index=train[train_feature].columns).sort_values(ascending=False)
```

```python
import numpy as np

pred_surv = estimator.predict_survival_function(temp[train_feature])
time_points = np.arange(1, 50)
for i, surv_func in enumerate(pred_surv[:3]):
    plt.step(time_points, surv_func(time_points), where="post",
             label="Sample %d" % (i + 1))
plt.ylabel("est. probability of survival $\hat{S}(t)$")
plt.xlabel("time $t$")
plt.legend(loc="best")
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.pipeline import Pipeline

pipe = Pipeline([('select', SelectKBest(fit_and_score_features, k=3)),
                 ('model', CoxPHSurvivalAnalysis())])
```

```python
train["是否转阴"] = p
train["住院天数截止"] = days
```

```python
train_list.append("是否转阴")
train_list.append("住院天数截止")
train[train_list].to_excel("Train.xlsx")
```

```python
from sklearn.model_selection import GridSearchCV, KFold

param_grid = {'select__k': np.arange(1, train[train_list[:-2]].shape[1] + 1)}
cv = KFold(n_splits=2, random_state=1, shuffle=True)
gcv = GridSearchCV(pipe, param_grid, return_train_score=True, cv=cv)
gcv.fit(train[train_list], new_data_y)

results = pd.DataFrame(gcv.cv_results_).sort_values(by='mean_test_score', ascending=False)
results.loc[:, ~results.columns.str.endswith("_time")]
```

```python
pipe.set_params(**gcv.best_params_)
pipe.fit(train[train_list], new_data_y)

encoder, transformer, final_estimator = [s[1] for s in pipe.steps]
pd.Series(final_estimator.coef_, index=encoder.encoded_columns_[transformer.get_support()])
```
